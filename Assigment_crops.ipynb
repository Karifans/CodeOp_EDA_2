```python
crops_main = [
    'Cereals, primary',
    'Wheat', 
    'Maize (corn)',
    'Rice', 
    'Soya beans',
    'Potatoes',
    'Oil palm fruit'

]

filter_crops = crops[crops['Item'].isin(crops_main)]
item_totals = filter_crops.groupby(['Item', 'Year'])['Value'].sum().reset_index()

plt.figure(figsize=(12, 6))

for crop in crops_main:
    crop_data = item_totals[item_totals['Item'] == crop]
 
    plt.plot(crop_data['Year'], crop_data['Value'], 
             label=crop, marker='o', linestyle='-', alpha=0.8)

plt.title("Main Crops Trend 2000-2022", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Bil in T", fontsize=12)

plt.legend(title="Crop", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
```
