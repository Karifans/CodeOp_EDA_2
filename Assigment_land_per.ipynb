```python
percentage_data = land_use[land_use['Unit'] == '%']

print(percentage_data[['Region', 'Value', 'Year','Unit', 'Item']].head())
```

```python
filter_data = percentage_data[
    (percentage_data['Region'] == 'South America') & 
    (percentage_data['Year'] == 2022)
]
mean_value = filter_data['Value'].mean()

print(filter_data[['Region', 'Value', 'Year', 'Unit', 'Item']].head(10))
```

```python
#Cropland
sum_data = percentage_data[percentage_data['Item'] == 'Cropland'].groupby(['Region', 'Year'])['Value'].mean().reset_index()

regions = sum_data['Region'].unique()

plt.figure(figsize=(12, 6))

for region in regions:

    region_data = sum_data[sum_data['Region'] == region]
    
    
    plt.plot(region_data['Year'], region_data['Value'], 
             label=region, marker='o', linestyle='-', alpha=0.8)


plt.title("Cropland Trends by Region from 2000-2022", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Land Use (%)", fontsize=12)


plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
```

```python
#Forest land
sum_data = percentage_data[percentage_data['Item'] == 'Forest land'].groupby(['Region', 'Year'])['Value'].mean().reset_index()

regions = sum_data['Region'].unique()

plt.figure(figsize=(12, 6))

for region in regions:

    region_data = sum_data[sum_data['Region'] == region]
    
    
    plt.plot(region_data['Year'], region_data['Value'], 
             label=region, marker='o', linestyle='-', alpha=0.8)


plt.title("Forest Land Trends by Region from 2000-2022", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Land Use (%)", fontsize=12)


plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
```

```python

cropland_land = land_use[(land_use['Item'] == 'Cropland') & 
                         (land_use['Unit'] == '%') & 
                         (land_use['World'] == 'World')]

forest_land = land_use[(land_use['Item'] == 'Forest land') & 
                       (land_use['Unit'] == '%') & 
                       (land_use['World'] == 'World')]

combined_land = pd.concat([cropland_land.assign(LandType='Cropland'),
                           forest_land.assign(LandType='Forest land')])

sns.catplot(data=combined_land, x='Year', y='Value', hue='LandType', kind='bar', height=6, aspect=2)

plt.title('Comparison of Cropland and Forest Land Use by World', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Land Use Percentage (%)', fontsize=12)

plt.xticks(rotation=90)

plt.show()
```
